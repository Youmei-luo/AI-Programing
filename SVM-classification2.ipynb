```python
import sys
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

plt.figure(figsize=(5,5))

# Two center points in binary classification
array_of_center_points=[(-1,-1),(0.5,0.5)]
# Create data set
array_of_coordinates_X,label_Y=make_blobs(n_samples=100,n_features=2,centers=array_of_center_points,
                       cluster_std=0.25,random_state=0)
# Data set is divided into learning data (coordinates), unknown data (coordinates),

learning_data_X,testdata_X,learning_data_label_Y,test_data_label_Y=train_test_split(array_of_coordinates_X,
                                                         label_Y,test_size=0.2,random_state=4)
# Learn with SVM
linearSVM=SVC(kernel='linear',C=sys.maxsize) # Strict misclassification
linearSVM.fit(learning_data_X,learning_data_label_Y) # Learning completed
# Draw decision boundaries
coefficient = linearSVM.coef_[0] # coef [0] is the coefficient of X and Y
slope = -coefficient[0] / coefficient[1] # This calculates the slope
x_drawing_point = np.linspace(-5, 5, 50)
y_drawing_point = slope * x_drawing_point - (linearSVM.intercept_[0]) / coefficient[1] # intercept_ is used for intercept
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
# Calculate margin coordinates
margin = 1/np.sqrt(np.sum(linearSVM.coef_**2))
margin_drawing_pointY_1 = y_drawing_point - np.sqrt(1+slope**2) * margin
margin_drawing_pointY_2 = y_drawing_point + np.sqrt(1+slope**2) * margin
# Draw margin
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_2, linestyle='dashed',color='k')
# Drawing training data on a scatter plot
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

plt.xlim(-2, 1.5) # Graph drawing range
plt.ylim(-2, 1.5)
plt.show()


plt.figure(figsize=(5,5))

# Drawing training data on a scatter plot
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_2, linestyle='dashed',color='k')
plt.scatter(linearSVM.support_vectors_[:, 0],linearSVM.support_vectors_[:, 1], 
            linewidth=2,edgecolors='k');

plt.xlim(-2, 1.5) # Graph drawing range
plt.ylim(-2, 1.5)
plt.show()

print(linearSVM.support_vectors_) # Support vector coordinates


test_data_classification_result=linearSVM.predict(testdata_X) # Obtain a label as a classification result to test data

plt.figure(figsize=(5,5))
# Drawing training data on a scatter plot
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Drawing unknown data on a scatter plot
plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_classification_result,cmap=cm.Accent,marker="x")

# Decision boundary, margin
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_2, linestyle='dashed',color='k')
plt.scatter(linearSVM.support_vectors_[:, 0],linearSVM.support_vectors_[:, 1], 
            linewidth=2,edgecolors='k');

plt.xlim(-2, 1.5) # Graph drawing range
plt.ylim(-2, 1.5)
plt.show()

print(test_data_classification_result)
print(test_data_label_Y)
print(accuracy_score(test_data_label_Y,test_data_classification_result)) # accuracy


# Two center points in binary classification (make get  closer two groupes than before)
array_of_center_points=[(-0.8,-0.9),(0.2,0.3)]

# The data set was recreated because the center point changed
array_of_coordinates_X,label_Y=make_blobs(n_samples=100,n_features=2,centers=array_of_center_points,cluster_std=0.37,random_state=0)

# Do over test data division
learning_data_X,testdata_X,learning_data_label_Y,test_data_label_Y=train_test_split(array_of_coordinates_X,
                                                         label_Y,test_size=0.2,random_state=4)

############################### Adding mean noise data to training data
learning_data_X=np.vstack((learning_data_X, [-0.05,-1]))
learning_data_label_Y=np.append(learning_data_label_Y,1)
############################### Added mean noise data to training data

# Start over learning
linearSVM=SVC(kernel='linear',C=sys.maxsize)
linearSVM.fit(learning_data_X,learning_data_label_Y) # learning
test_data_classification_result=linearSVM.predict(testdata_X) # Obtain a label as a classification result to test data

# Recalculate decision boundaries
coefficient = linearSVM.coef_[0] # coef [0] is the coefficient of X and Y
slope = -coefficient[0] / coefficient[1] # This calculates the slope
x_drawing_point = np.linspace(-5, 5)
y_drawing_point = slope * x_drawing_point - (linearSVM.intercept_[0]) / coefficient[1] # intercept_ is used for intercept(error)

# Recalculate margin
margin = 1/np.sqrt(np.sum(linearSVM.coef_**2))
margin_drawing_pointY_1 = y_drawing_point - np.sqrt(1+slope**2) * margin
margin_drawing_pointY_2 = y_drawing_point + np.sqrt(1+slope**2) * margin

plt.figure(figsize=(5,5))
# Drawing training data on a scatter plot
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Drawing unknown data on a scatter plot
#plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_classification_result,cmap=cm.Accent,marker="x")

# Decision boundary、margin
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_2, linestyle='dashed',color='k')

plt.xlim(-2, 1.5) # Graph drawing range
plt.ylim(-2, 1.5)
plt.show()

print(test_data_classification_result)
print(test_data_label_Y)
print(accuracy_score(test_data_label_Y,test_data_classification_result)) # Accuracy


# Re-learn with lower penalty C
linearSVM=SVC(kernel='linear',C=5) # make smaller penalty C
linearSVM.fit(learning_data_X,learning_data_label_Y) # learning
test_data_classification_result=linearSVM.predict(testdata_X) # Obtain a label as a classification result to test data

# Recalculate decision boundaries
coefficient = linearSVM.coef_[0] # coef [0] is the coefficient of X and Y
slope = -coefficient[0] / coefficient[1] # This calculates the slope
x_drawing_point = np.linspace(-5, 5)
y_drawing_point = slope * x_drawing_point - (linearSVM.intercept_[0]) / coefficient[1] # intercept_ is used for intercept(error)

# Recalculate margin
margin = 1/np.sqrt(np.sum(linearSVM.coef_**2))
margin_drawing_pointY_1 = y_drawing_point - np.sqrt(1+slope**2) * margin
margin_drawing_pointY_2 = y_drawing_point + np.sqrt(1+slope**2) * margin

plt.figure(figsize=(5,5))
# Learning data
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Unknown data
plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_classification_result,cmap=cm.Accent,marker="x")

# Decision boundary、margin
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, margin_drawing_pointY_2, linestyle='dashed',color='k')

plt.xlim(-2, 1.5) # Graph drawing range
plt.ylim(-2, 1.5)
plt.show()

print(test_data_classification_result)
print(test_data_label_Y)
print(accuracy_score(test_data_label_Y,test_data_classification_result)) # Accuracy


# Kernel method
from sklearn.datasets import make_gaussian_quantiles
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons

array_of_coordinates_X,label_Y=make_gaussian_quantiles(n_features=2,n_classes=2,n_samples=300,random_state=0)
learning_data_X,testdata_X,learning_data_label_Y,test_data_label_Y=train_test_split(array_of_coordinates_X,
                                                         label_Y,test_size=0.2,random_state=0)

non_linearSVM=SVC(kernel='rbf') # Gaussian kernel
non_linearSVM.fit(learning_data_X,learning_data_label_Y)
test_data_classification_result=non_linearSVM.predict(testdata_X)

plt.figure(figsize=(5,5))
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,cmap=cm.Accent,marker="o")
plt.scatter(testdata_X[:,0],testdata_X[:,1],c=test_data_classification_result,cmap=cm.Accent,marker="x")
plt.xlim(-3.5, 3.5) # Graph drawing range
plt.ylim(-3.5, 3.5)
plt.show()

print(accuracy_score(test_data_label_Y,test_data_classification_result)) # Accuracy

X, y = make_circles(n_samples=300,
                  shuffle = True,
                  noise = 0.2,
                  random_state=None,
                  factor = 0.7)
print(X)
print(y)
 
a0, b0 = X[y==0,0], X[y==0,1]
a1, b1 = X[y==1,0], X[y==1,1]
 
plt.figure(figsize=(8, 7))
 
plt.scatter(a0, b0, marker='o', s=25, label="y = 0")
plt.scatter(a1, b1, marker='o', s=25, label="y = 1")
plt.legend()
plt.xlabel("a")
plt.ylabel("b")
plt.show()

X, y = make_moons(n_samples=300,
                  shuffle = True,
                  noise = None,
                  random_state = True,)
print(X)
print(y)
 
a0, b0 = X[y==0,0], X[y==0,1]
a1, b1 = X[y==1,0], X[y==1,1]
 
plt.figure(figsize=(8, 7))
 
plt.scatter(a0, b0, marker='o', s=25, label="y = 0")
plt.scatter(a1, b1, marker='o', s=25, label="y = 1")
plt.legend()
plt.xlabel("a")
plt.ylabel("b")
plt.show()
```
