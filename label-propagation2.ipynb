```python
import seaborn as sns
iris_data = sns.load_dataset("iris")
explanatory_variable_x=iris_data.values[:,[0,1,2,3]]
explanatory_variable_x=explanatory_variable_x[:,:2] # Extract sepal_length and sepal_width

y=iris_data.values[:,4]
y = [s.replace('setosa', '0') for s in y] # setosa is 0
y = [s.replace('versicolor', '1') for s in y] # versicolor is 1
y = [s.replace('virginica', '2') for s in y] # virginica is 2
y = [int(n) for n in y] # Change characters to numbers


import numpy as np
objective_variable_y = np.array(y) # NumPy's array

import matplotlib.pyplot as plt
plt.figure()
plt.scatter(explanatory_variable_x[objective_variable_y==0,0],explanatory_variable_x[objective_variable_y==0,1],c = 'red')
plt.scatter(explanatory_variable_x[objective_variable_y==1,0],explanatory_variable_x[objective_variable_y==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[objective_variable_y==2,0],explanatory_variable_x[objective_variable_y==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[objective_variable_y==-1,0],explanatory_variable_x[objective_variable_y==-1,1],c = 'black')

lower_limit_of_range=0
upper_limit_of_range=objective_variable_y.size
randomize_rate=0.7
number_to_randomize=int((objective_variable_y.size)*randomize_rate)
index_without_label=np.random.choice(range(lower_limit_of_range,upper_limit_of_range),
                              number_to_randomize,replace=False)
label_array=np.copy(objective_variable_y)
label_array[index_without_label] = -1

plt.figure()
plt.scatter(explanatory_variable_x[label_array==0,0],explanatory_variable_x[label_array==0,1],c = 'red')
plt.scatter(explanatory_variable_x[label_array==1,0],explanatory_variable_x[label_array==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[label_array==2,0],explanatory_variable_x[label_array==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[label_array==-1,0],explanatory_variable_x[label_array==-1,1],c = 'black')

from sklearn.semi_supervised import LabelPropagation
label_propagation=LabelPropagation(max_iter=10000) # Maximum number of repetitions
label_propagation.fit(explanatory_variable_x,label_array) # Learning
Predicted_y=label_propagation.predict(explanatory_variable_x)
print(Predicted_y)

plt.figure()
plt.scatter(explanatory_variable_x[Predicted_y==0,0],explanatory_variable_x[Predicted_y==0,1],c = 'red')
plt.scatter(explanatory_variable_x[Predicted_y==1,0],explanatory_variable_x[Predicted_y==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[Predicted_y==2,0],explanatory_variable_x[Predicted_y==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[Predicted_y==-1,0],explanatory_variable_x[Predicted_y==-1,1],c = 'black')
plt.show()

# Compare at only the unlabeled parts, and score
print(label_propagation.score(explanatory_variable_x[index_without_label],objective_variable_y[index_without_label]))

# 
# Read 4 columns again from the original data
explanatory_variable_x=iris_data.values[:,[0,1,2,3]]
print(explanatory_variable_x)

# Do over[Without label]
lower_limit_of_range=0
upper_limit_of_range=objective_variable_y.size
randomize_rate=0.7
number_to_randomize=int((objective_variable_y.size)*randomize_rate)
index_without_label=np.random.choice(range(lower_limit_of_range,upper_limit_of_range),
                              number_to_randomize,replace=False)
label_array=np.copy(objective_variable_y)
label_array[index_without_label] = -1

label_propagation=LabelPropagation(max_iter=10000) # Maximum number of repetitions
label_propagation.fit(explanatory_variable_x,label_array) # Learn using 4D explanatory variables
Predicted_y=label_propagation.predict(explanatory_variable_x)
print(Predicted_y)

plt.figure()
plt.scatter(explanatory_variable_x[Predicted_y==0,0],explanatory_variable_x[Predicted_y==0,1],c = 'red')
plt.scatter(explanatory_variable_x[Predicted_y==1,0],explanatory_variable_x[Predicted_y==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[Predicted_y==2,0],explanatory_variable_x[Predicted_y==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[Predicted_y==-1,0],explanatory_variable_x[Predicted_y==-1,1],c = 'black')
plt.show()

print(label_propagation.score(explanatory_variable_x[index_without_label],objective_variable_y[index_without_label]))

# Variable [randomize_rate] was changed
lower_limit_of_range=0
upper_limit_of_range=objective_variable_y.size
randomize_rate=0.9 ######### This was changed!
number_to_randomize=int((objective_variable_y.size)*randomize_rate)
index_without_label=np.random.choice(range(lower_limit_of_range,upper_limit_of_range),
                              number_to_randomize,replace=False)
label_array=np.copy(objective_variable_y)
label_array[index_without_label] = -1

label_propagation=LabelPropagation(max_iter=10000) # Maximum number of repetitions
label_propagation.fit(explanatory_variable_x,label_array) # Learn using 4D explanatory variables
Predicted_y=label_propagation.predict(explanatory_variable_x)
print(Predicted_y)

plt.figure()
plt.scatter(explanatory_variable_x[Predicted_y==0,0],explanatory_variable_x[Predicted_y==0,1],c = 'red')
plt.scatter(explanatory_variable_x[Predicted_y==1,0],explanatory_variable_x[Predicted_y==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[Predicted_y==2,0],explanatory_variable_x[Predicted_y==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[Predicted_y==-1,0],explanatory_variable_x[Predicted_y==-1,1],c = 'black')
plt.show()

print(label_propagation.score(explanatory_variable_x[index_without_label],objective_variable_y[index_without_label]))

# Variable [randomize_rate] was changed
lower_limit_of_range=0
upper_limit_of_range=objective_variable_y.size
randomize_rate=0.9 ######### This was changed!
number_to_randomize=int((objective_variable_y.size)*randomize_rate)
index_without_label=np.random.choice(range(lower_limit_of_range,upper_limit_of_range),
                              number_to_randomize,replace=False)
label_array=np.copy(objective_variable_y)
label_array[index_without_label] = -1

label_propagation=LabelPropagation(max_iter=10000) # Maximum number of repetitions
label_propagation.fit(explanatory_variable_x,label_array) # Learn using 4D explanatory variables
Predicted_y=label_propagation.predict(explanatory_variable_x)
print(Predicted_y)

plt.figure()
plt.scatter(explanatory_variable_x[Predicted_y==0,0],explanatory_variable_x[Predicted_y==0,1],c = 'red')
plt.scatter(explanatory_variable_x[Predicted_y==1,0],explanatory_variable_x[Predicted_y==1,1],c = 'blue')
plt.scatter(explanatory_variable_x[Predicted_y==2,0],explanatory_variable_x[Predicted_y==2,1],c = 'yellow')
plt.scatter(explanatory_variable_x[Predicted_y==-1,0],explanatory_variable_x[Predicted_y==-1,1],c = 'black')
plt.show()

print(label_propagation.score(explanatory_variable_x[index_without_label],objective_variable_y[index_without_label]))

lower_limit_of_rate = 0.05
upper_limit_of_rate = 0.45
step = 0.05
array_of_labeled_rate = np.arange(lower_limit_of_rate,upper_limit_of_rate,step)
print(array_of_labeled_rate)

def make_unlabeled_index(percentage_to_be_labeled):
    lower_limit_of_range=0
    upper_limit_of_range=objective_variable_y.size
    randomize_rate=1-percentage_to_be_labeled 
    number_to_randomize=int((objective_variable_y.size)*randomize_rate)
    index_without_label=np.random.choice(range(lower_limit_of_range,upper_limit_of_range),
                                  number_to_randomize,replace=False)
    return index_without_label

overall_size=objective_variable_y.size
number_of_trials=100
average_accuracy_rate = []
average_accuracy_rate.append(0.00)
for labeled_percentage in array_of_labeled_rate:
    accuracy_rate_total=0
    for i in range(number_of_trials):
        index_without_label=make_unlabeled_index(labeled_percentage)
        label_array=np.copy(objective_variable_y)
        label_array[index_without_label]= -1
        label_propagation.fit(explanatory_variable_x,label_array)
        accuracy_rate_total += label_propagation.score(explanatory_variable_x[index_without_label],
                             objective_variable_y[index_without_label])    
    print("Labeled percentage :",'{:.2f}'.format(labeled_percentage),end="")
    print(", Average accuracy rate:",'{:.6f}'.format(accuracy_rate_total/number_of_trials))
    average_accuracy_rate.append(accuracy_rate_total/number_of_trials)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,0.45,0.05)
plt.plot(x,average_accuracy_rate)
plt.show()       

for index in range(len(iris_data)):
    print (iris_data.species[index], Predicted_y[index])

    
```
