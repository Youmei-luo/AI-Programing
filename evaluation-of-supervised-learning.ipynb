```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True,precision=20) # Do not use exponential notation. 20 decimal places

medical_data=load_breast_cancer() # 
pd.set_option('display.max_columns', 40)
data_frame = pd.DataFrame(medical_data.data, columns=medical_data.feature_names)
data_frame # It can be displayed without using print function

explanatory_variableX=medical_data.data
objective_variabley=1-medical_data.target # Swap positive (0) and negative (1) by subtraction

# Take  two explanatory variables randomly and draw a scatter plot
plt.figure(figsize = (8, 8)) # Preparation for displaying
plt.scatter(explanatory_variableX[:,0],explanatory_variableX[:,2]) # Mean radius, mean perimeter
plt.show()

explanatory_variableX=explanatory_variableX[:,:13] # Extract only (area error) and make simple regression

logistic_regression=LogisticRegression()
logistic_regression.fit(explanatory_variableX,objective_variabley)
estimated_y=logistic_regression.predict(explanatory_variableX)
print(estimated_y) # Estimated y
print(objective_variabley) # Prepared y

# from sklearn.metrics import. Function of accuracy_score 
print(accuracy_score(objective_variabley,estimated_y))

cm=confusion_matrix(objective_variabley,estimated_y)
an =(cm[1][1] + cm[0][0]) / (cm[1][1] + cm[0][0] + cm[0][1] + cm[1][0])
print(an)

# from sklearn.metrics import. Function of precision_score
print(precision_score(objective_variabley,estimated_y))

an =cm[1][1] / (cm[1][1] + cm[0][1])
print(an)


# from sklearn.metrics import. Function of recall_score
print(recall_score(objective_variabley,estimated_y))

an =cm[1][1] / (cm[1][1] + cm[1][0])
print(an)

cm=confusion_matrix(objective_variabley,estimated_y)
print(cm)

# from sklearn.metrics import. Function of  f1_score
print(f1_score(objective_variabley,estimated_y))

np.set_printoptions(threshold=3000) # 
print(logistic_regression.predict_proba(explanatory_variableX))

re_predictedy=(logistic_regression.predict_proba(explanatory_variableX)[:,1]>0.3) # Limit  0.3
print(confusion_matrix(objective_variabley,re_predictedy))
print(accuracy_score(objective_variabley,re_predictedy))
print(recall_score(objective_variabley,re_predictedy))

predicted_probability=logistic_regression.predict_proba(explanatory_variableX) # predicted_probability
# Correct label and predicted probability are arguments
false_positive_rate_FPR,true_positive_rate_TPR,limit=roc_curve(objective_variabley,predicted_probability[:,1]) 
plt.figure(figsize = (8, 8))
plt.step(false_positive_rate_FPR,true_positive_rate_TPR,color='b')
plt.fill_between(false_positive_rate_FPR,true_positive_rate_TPR,0,color='b',alpha=0.5)
plt.xlabel('FPR') # False positive rate FP/(TN+FP) 
plt.ylabel('TPR') # True positive rate TP/(TP+FN) 
plt.show()

ROC_display_data_frame=pd.DataFrame({'Limit':limit,'False positive rate FPR': false_positive_rate_FPR, 
                            'True positive rate TPR': true_positive_rate_TPR})
display(ROC_display_data_frame.iloc[:,[2,0,1]]) # Swap display order

# from sklearn.metrics import. Function of  roc_auc_score
print(roc_auc_score(objective_variabley,predicted_probability[:,1]))
```
