```python
#Learn the calculation procedure of neural network (simple perceptron)
class how_to_buy_snacks(object):
    def __init__ (self,the_name_of_how_to_buy,list_of_purchasing,correct_answer): # Constructor 
        self.the_name_of_how_to_buy=the_name_of_how_to_buy
        self.list_of_purchasing=list_of_purchasing # [Buy : 1, Not by : 0]
        self.correct_answer=correct_answer # [NG buying : 1, OK buying : 0]

class parts(object):
    def __init__ (self,parts_name,a_value_inside): # Constructor
        self.parts_name=parts_name
        self.a_value_inside=a_value_inside
    def increase(self,a_value_to_increase):   # Add the specified value
        self.a_value_inside=self.a_value_inside+a_value_to_increase
    def decrease(self,a_value_to_decrease):   # Subtract specified value
        self.a_value_inside=self.a_value_inside-a_value_to_decrease

# 3 parts B,C,and D and  value inside
list_of_snack_parts=[parts("partsB",1), parts("partsC",3), parts("partsD",8)] 
partsA=parts("partsA",6)
#Number of snacks is same as number of list_of_snack_parts
number_of_snacks=len(list_of_snack_parts) 

list_of_training_data=[ # 5 types training data
    how_to_buy_snacks("how_to_buy_1",[1,1,1],"NG"),
    how_to_buy_snacks("how_to_buy_2",[1,1,0],"NG"),
    how_to_buy_snacks("how_to_buy_3",[1,0,1],"OK"),
    how_to_buy_snacks("how_to_buy_4",[0,1,1],"OK"),
    how_to_buy_snacks("how_to_buy_5",[1,0,0],"OK")]

list_of_unknown_data=[ # 3 types unknown data
    how_to_buy_snacks("how_to_buy_6",[0,1,0],"OK"),
    how_to_buy_snacks("how_to_buy_7",[0,0,1],"OK"),
    how_to_buy_snacks("how_to_buy_8",[0,0,0],"OK")]

def judge_how_to_buy(data_how_to_buy,partsA): # Parts B, C, D, and A judge OK or NG for how to buy
    sum=0
    for order in range(number_of_snacks):
        sum=sum+data_how_to_buy.list_of_purchasing[order]*list_of_snack_parts[order].a_value_inside

    print(str(sum)+" > " + str(partsA.a_value_inside)+" ", end="") # Display comparison formula
    
    if sum > partsA.a_value_inside : 
        return "NG" # Judgment is NG, part A gets excited
    else:
        return "OK" # Judgment is NG

def learn_with_given_data(data_how_to_buy,partsA):
    print(data_how_to_buy.the_name_of_how_to_buy+" is ",end="")
    print(",".join([str(x) for x in data_how_to_buy.list_of_purchasing])+",",end="")
    judgment_result=judge_how_to_buy(data_how_to_buy,partsA)
    print(" Judgment by this comparison is "+judgment_result+",",end="")

    # Increase or decrease the value in BCDA based on the judgment result and the correct answer
    if data_how_to_buy.correct_answer == "OK" and judgment_result == "NG": # NG type 1
        # Decrease BCD by 1 and increase A by 1
        for order in range(number_of_snacks): # order is 0,1,2
            if data_how_to_buy.list_of_purchasing[order]==1: # When this part buys a snack
                list_of_snack_parts[order].decrease(1)
        partsA.increase(1)            
    if data_how_to_buy.correct_answer == "NG" and judgment_result == "OK": # NG type 2
        # Increase BCD by 1 and decrease A by 1
        for order in range(number_of_snacks): # order is 0,1,2
            if data_how_to_buy.list_of_purchasing[order]==1: # When this part buys a snack
                list_of_snack_parts[order].increase(1)
        partsA.decrease(1)
        
    print("The answer is "+data_how_to_buy.correct_answer+",As a result, BCDA are ",end="")    
    print(",".join([str(x.a_value_inside) for x in list_of_snack_parts]),end="")
    print(","+str(partsA.a_value_inside)) # The value in A is also output


# 
list_of_snack_parts[0].a_value_inside=1
list_of_snack_parts[1].a_value_inside=3
list_of_snack_parts[2].a_value_inside=8
partsA.a_value_inside=6

# Learn repeatedlyâ€¦ After execution, BCDA should be 4,5,2,7
for i in range(1,7): # The variable i  is from 1 to 6
    for training_data in (list_of_training_data): # 5 types training data
        learn_with_given_data(training_data,partsA)
    print("=== The "+str(i)+"th learning is the end ===") 
```
