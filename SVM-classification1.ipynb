```python
import sys
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Two center points in binary classification
array_of_center_points=[(-1,-1),(0.5,0.5)]

# Create data set
array_of_coordinates_X,label_Y=make_blobs(n_samples=100,n_features=2,centers=array_of_center_points,
                       cluster_std=0.25,random_state=0)

# Data set is divided into learning data (coordinates), unknown data (coordinates), 
#       learning data (labels), and unknown data (labels)
learning_data_X,testdata_data_X,learning_data_label_Y,test_data_label_Y=train_test_split(array_of_coordinates_X,
                                                         label_Y,test_size=0.2,random_state=4)
plt.figure(figsize=(5,5))
# Create scatter plot with training data
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_data_X[:,0],testdata_data_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

plt.xlim(-2, 1.5) # Graph plot range
plt.ylim(-2, 1.5)
plt.show()

print(array_of_coordinates_X)
print(label_Y)

i = 0
j = 0
for count in label_Y:
    if count == 0:
        i = i + 1
    else:
        j = j + 1
    
print(i)
print(j)
print(len(test_data_label_Y))


# Draw decision boundaries appropriately

plt.figure(figsize=(5,5))
# Learning data
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_data_X[:,0],testdata_data_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

x_range_of_decision_boundary=np.linspace(-2,1.5,50) # Get 50 numbers at equal intervals
plt.plot(x_range_of_decision_boundary,-0.01*x_range_of_decision_boundary-0.24,'k') # Decision boundaries 1
plt.plot(x_range_of_decision_boundary,-0.15*x_range_of_decision_boundary-0.24,'k') # Decision boundaries 2

plt.xlim(-2, 1.5) # Graph plot range
plt.ylim(-2, 1.5)
plt.show()


plt.figure(figsize=(5,5))

# Create scatter plot with training data
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_data_X[:,0],testdata_data_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

# Learning with SVM
linear_SVM=SVC(kernel='linear',C=sys.maxsize) # Strict misclassification
linear_SVM.fit(learning_data_X,learning_data_label_Y) # Learning completed

# Draw decision boundaries
coefficient = linear_SVM.coef_[0] # coef [0] is the coefficient of X and Y
slope = -coefficient[0] / coefficient[1] # This calculates the slope
x_drawing_point = np.linspace(-5, 5, 50)
y_drawing_point = slope * x_drawing_point - (linear_SVM.intercept_[0]) / coefficient[1] # intercept is used for sections
plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')

plt.xlim(-2, 1.5) # Graph plot range
plt.ylim(-2, 1.5)
plt.show()

print(linear_SVM.coef_) # coefficient
print(linear_SVM.intercept_) # Constant term


plt.figure(figsize=(5,5))

# Create scatter plot with training data
plt.scatter(learning_data_X[:,0],learning_data_X[:,1],c=learning_data_label_Y,marker="o",s=10,cmap=cm.Accent)
# Do not display unknown data yet
#plt.scatter(testdata_data_X[:,0],testdata_data_X[:,1],c=test_data_label_Y,cmap=cm.seismic,marker="o")

# Calculate margin coordinates
margin = 1/np.sqrt(np.sum(linear_SVM.coef_**2))
y_of_margin_drawing_point_1 = y_drawing_point - np.sqrt(1+slope**2) * margin
y_of_margin_drawing_point_2 = y_drawing_point + np.sqrt(1+slope**2) * margin

plt.plot(x_drawing_point, y_drawing_point,linestyle='solid',color='k')
plt.plot(x_drawing_point, y_of_margin_drawing_point_1, linestyle='dashed',color='k')
plt.plot(x_drawing_point, y_of_margin_drawing_point_2, linestyle='dashed',color='k')
# Only one dashed line in the margin area is drawn 

plt.xlim(-2, 1.5) # Graph plot range
plt.ylim(-2, 1.5)
plt.show()

print(margin)
print(y_of_margin_drawing_point_1,y_of_margin_drawing_point_2)


```
